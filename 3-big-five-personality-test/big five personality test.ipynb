### Modelling

#### Drop country columns as there is no correlation

```python
df.drop(["country", "country_enc"], axis=1, inplace=True)
df.head()
```

#### scale all values to between 0 and 1 for easier calculation

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(0,1))
columns = list(df.columns)
df = pd.DataFrame(scaler.fit_transform(df), columns=columns)
df.head()
```

#### Visualise best amount of clusters

```python
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer

kmeans = KMeans()
visualizer = KElbowVisualizer(kmeans, k=(2,15))
visualizer.fit(df)
visualizer.poof()
```

#### Fit the KMeans model

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=5)
k_fit = kmeans.fit(df)
predictions = k_fit.labels_
df['clusters'] = predictions
df.head(8)
```

```python
df["clusters"].value_counts()
```

#### Find a pattern in the clusters
This is done by taking the average of each type of attribute then grouping by the cluster and taking the average of that

```python
ext_cols = [f"EXT{i}" for i in range(1, 11, 1)]
est_cols = [f"EST{i}" for i in range(1, 11, 1)]
arg_cols = [f"AGR{i}" for i in range(1, 11, 1)]
csn_cols = [f"CSN{i}" for i in range(1, 11, 1)]
opn_cols = [f"OPN{i}" for i in range(1, 11, 1)]

df_averages = pd.DataFrame()

df_averages["EXT_avg"] = df[ext_cols].mean(axis=1)
df_averages["EST_avg"] = df[est_cols].mean(axis=1)
df_averages["AGR_avg"] = df[arg_cols].mean(axis=1)
df_averages["CSN_avg"] = df[csn_cols].mean(axis=1)
df_averages["OPN_avg"] = df[opn_cols].mean(axis=1)
df_averages['clusters'] = predictions

df_averages.groupby('clusters').mean()
```

#### Visualizing the patterns for each cluster

```python
plt.figure(figsize=(22,3))

dataclusters = df_averages.groupby('clusters').mean()
for i in range(0, 5):
    plt.subplot(1,5,i+1)
    plt.bar(dataclusters.columns, dataclusters.iloc[i], color='green', alpha=0.2)
    plt.plot(dataclusters.columns, dataclusters.iloc[i], color='red')
    plt.title('Cluster ' + str(i))
    plt.xticks(rotation=45)
    plt.ylim(0,1);
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca_fit = pca.fit_transform(df)

df_pca = pd.DataFrame(data=pca_fit, columns=['PCA1', 'PCA2'])
df_pca['clusters'] = predictions
df_pca.head()
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(data=df_pca, x='PCA1', y='PCA2', hue='clusters', palette='Set2', alpha=0.8)
plt.title('Personality Clusters after PCA');
```
